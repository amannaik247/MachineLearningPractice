```python
import numpy as np
import pandas as pd
```

```python
import math
```

```python
# Generate random integers between 0 and 100 for a 8x5 array
data = np.random.randint(0, 100, size=(8, 5))

# Create DataFrame with specified column names
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

df.head()
```

```python
class StandardScalerDF:
  """
  fit_transform method similar to standardscaler but instead of returning an np.array
  we return the same dataframe with same column names
  """
  # def __init__(self, df):
  #   self.df = df

  def fit_transform(self, df):
    # loop over each column
    self.rows = df.shape[0]
    self.cols = df.shape[1]
    self.df_new = pd.DataFrame()
    for col in range(self.cols):

      # get mean
      self.mean = df.iloc[:, col].sum() / self.rows

      # get squared distances
      self.l2dist = df.iloc[:, col].apply(lambda x: (x-self.mean)**2)

      # get sum of all distances
      self.var = self.l2dist.sum() / self.rows

      # Divide by total rows and take root
      self.std = math.sqrt(self.var)

      # transform formula z = (x âˆ’ mean) / std
      self.df_new[df.columns[col]] = df.iloc[:, col].apply(lambda x: (x - self.mean) / self.std)
    
    return self.df_new
```

```python
scaler = StandardScalerDF()

df_new = scaler.fit_transform(df)

df_new
```
